```python
# --------------- Imports ---------------
import random
from random import choice
from time import sleep
import os
from character import You, TwinBrother # import characters for fighting scene in first world from character file
from weapon import pan, sword, fists # import weapons for fighting scene in first world from weapon file

# --------------- Movieportal game ---------------
class movieportal_game():
    def __init__(self, name:str, points: int):
        # Game is not over
        self.game_over = False
        # Defining self.points
        self.points = points
        
        # User starts with no portal pieces (empty list)
        self.portalpieces = []
        # Describing portal piece of first world
        first_portal_piece = str("Portal piece of first world")
        # Defining self.first_portal_piece
        self.first_portal_piece = first_portal_piece
        # Describing portal piece of second world
        second_portal_piece = str("Portal piece of second world")
        # Defining self.second_portal_piece
        self.second_portal_piece = second_portal_piece
        # Describing portal piece of third world
        third_portal_piece = str("Portal piece of third world")
        # Defining self.third_portal_piece
        self.third_portal_piece = third_portal_piece
        # Describing portal piece of first world
        fourth_portal_piece = str("Portal piece of fourth world")
        # Defining self.fourth_portal_piece
        self.fourth_portal_piece = fourth_portal_piece
        
        # User starts with no items (empty inventory list)
        items = []
        # Defining self.items
        self.items = items
        # Description of item key
        key = str("A key")
        # Defining self.key
        self.key = key
        # Description of item satchel
        satchel = str("The stolen satchel of the queen")
        # Defining self.satchel
        self.satchel = satchel
    
        # Welcoming player
        print("Indiana Jones: 'Welcome traveller, my name is Indiana Jones.'")
        # Ask the user for name, greet him/her and ask for help
        self.name = input("Indiana Jones: 'What is your name?'")
        
        # Users characteristics for fighting game in the first world (has to be after self.name is defined)
        you = You(name=self.name, health=100)
        # Charachteristics of twin brother for fighting game in the first world
        twinbrother = TwinBrother(name="Man with eyepatch", health=100, weapon=sword)
        
        # Dialogue pause
        sleep(1)
        # Dialogue
        print(f"Nice to meet you {self.name}!")
        print("Indiana Jones: 'The craziest thing has happened to me.'")
        print("Indiana Jones: 'I was running towards the temple of doom and all of a sudden a portal appeared in my pathway.'")
        # Dialogue pause
        sleep(1)
        # Dialogue
        print("Indiana Jones: 'I got sucked into this portal and now I'm stuck here...'")
        # Dialogue pause
        sleep(1)
        # Go to next scene
        self.helpout()
        
    # Function to create decision-point: (not) helping Indiana Jones
    def helpout(self):
        # List of answers
        answers = ["y", "n"]
        # Pre-define user input
        user_input = ""
        # Loop until the user gives a valid answer for the game to continue
        while user_input not in answers:
            # User types answer
            user_input = input("Indiana Jones: 'It seems like another dimension. Could you help me find my way back to the temple of doom?' (y/n)")
            # If answer is yes
            if user_input == "y":
                # Dialogue and dialogue pauses
                print("Thank you for wanting to help me. Let's start this adventure!")
                sleep(1)
                print("A bright blue portal swirls in front of you...")
                sleep(1)
                print("You step closer, feeling a strange pull.")
                # Calls the next scene
                self.portalscene()
            # If answer is no
            elif user_input == "n":
                # Dialogue
                print("I understand, it is a big job to ask of anyone. Maybe another time!")
                # Stops the game
                self.game_over = True
                # Restarts the game
                if __name__ == "__main__":
                    game = movieportal_game(name="Player", points=0)
                    game.start_game()
            # If input of the user isn't contained in the options, user should enter another option
            else:
                print("Please enter a valid answer.")

    # Function to choose a portal
    def portalscene(self):
        # Specifies variable that user needs all portal pieces of different world
        required_pieces = {self.first_portal_piece, self.second_portal_piece, self.third_portal_piece}
        # If user has achieved all portal pieces (aka all required portal pieces are contained in the set)
        if set(self.portalpieces) >= required_pieces:
            # Dialogue and dialogue pauses
            print("Hoorah, you've achieved all portal pieces! \nYou combine all portal pieces and a blue portal forms to the world of Indiana Jones.")
            sleep(2)
            # Refers to user name
            print(f"Indiana Jones: 'Thank you so much for helping me {self.name}! Now I can go back to my world.")
            sleep(2)
            print("Indiana Jones tips his hat and steps into the portal. The portal dissapears.")
            sleep(1)
            print("Congratulations! You've won the game.")
            # List of answers
            answers = ["y","n"]
            # Pre-define variable user input
            user_input = ""
            while user_input not in answers:
                # User types answer
                user_input = input("Do you want to play again? (y/n)")
                # If answer is yes
                if user_input == "y":
                    # Removes all elements (portal pieces) from list of achieved portal pieces
                    self.portalpieces.clear()
                # If answer is no
                elif user_input == "n":
                    # Dialogue
                    print("Thank you for playing this game. Till next time!")
                    # Stops the game
                    self.game_over = True
                    # Restarts the game
                    if __name__ == "__main__":
                        game = movieportal_game(name="Player", points=0)
                        game.start_game()
                # If input of the user isn't contained in the answers, user should enter another answer
                else:
                    print("Please enter a valid option for this game.")         
        # If not, user continues with choosing a world
        else:
            sleep(1)
        # Dialogue and dialogue pauses 
        print("\nYou feel light as air and open your eyes. There is no ground beneath you, yet you are standing. It seems like you are somewhere in space. The space is filled with all kinds of colours swirling around you. It is absolutely beautiful.")
        sleep(2)
        print("While you are taking the space in with eyes full of wonder, Indiana Jones clears his throat.")
        sleep(1)
        print("Indiana Jones: 'Since I've been here awhile already, I'll tell you what I know. \nAs you can see, there are four portals in front of us. \nEach portal belongs to a different movie dimension. \nTo get me back to the temple of doom we're going to have to build a portal to my world. \nAfter completing each portal adventure, we'll receive a piece of my dimensions portal. \nTo complete a world we'll need a certain amount of points. \nIn order to get me back to my world we'll need all portal pieces. I haven't succeeded to complete any portal adventures, but I guess you already figured that out, since I'm still here.'")
        sleep(4)
        print("Indiana Jones: 'There are four portals. Starting from the left, which one of the portals do you want to go into?'")
        # List of options
        portals = ["first", "second", "third", "fourth"]
        # Pre-define variable user input
        user_input = ""
        # Loop until the user gives a valid answer for the game to continue
        while user_input not in portals:
            # Specify user input: user can choose a portal
            user_input = input("Choose a portal: first/second/third/fourth")
            # If user selects first world, then call the first function of that world
            if user_input == "first":
                print("Here we go!")
                self.first_world()
            # If user selects second world, then call the first function of that world
            elif user_input == "second":
                print("OK, let's go!")
                self.second_world()
            # If user selects third world, then call the first function of that world
            elif user_input == "third":
                print("Let's hope for the best!")
                self.third_world()
            # If user selects fourth world, then call the first function of that world
            elif user_input == "fourth":
                print("Let's go!")
                self.fourth_world()
            # If input of the user isn't contained in the options, user should enter another option
            else:
                print("Please enter a valid option for this game.")

    # --------------- First world: Rapunzel ---------------
        
    # Beginning scene of Rapunzels world
    def first_world(self):
        # If the user has already achieved the first portal piece
        if self.first_portal_piece in self.portalpieces:
            # Pre-define user variable
            user_input = ""
            # List of answers
            answers = ["y", "n"]
            # Loop until the user enters one of the given options
            while user_input not in answers:
                # User types answer
                user_input = input("You've already completed this world. If you start again, you'll need to earn this portal piece again. Do you want to continue? (y/n)")
                # If answer is yes
                if user_input == "y":
                    # Remove the first portal piece from the list of portal pieces
                    self.portalpieces.remove(self.first_portal_piece)
                # If answer is no
                elif user_input == "n":
                    # User goes back to portal scene
                    self.portalscene()
                # If input of the user isn't contained in the options, user should enter another answer
                else:
                    print("Please enter a valid option.")
        # Start world with zero points for the user
        self.points = 0
        # Dialogue and dialogue pauses
        print(f"Welcome {self.name}, in order to win a portal piece and complete this world, you'll need to earn 100 points. Good luck!")
        sleep(2)
        print("\nYou find yourself in a small and round room. The walls are made out of stone and covered with beautiful paintings.")
        sleep(1)
        print("There's a narrow window high above you. There is a girl there brushing her extremely long and golden hair. \nHer back is turned towards you.")
        sleep(1)
        # List of options
        options = ["1","2", "3"]
        # Pre-define variable user input
        user_input = ""
        # Loop until the user enters one of the given options
        while user_input not in options:
            print("1. Say hello \n2. Walk towards the paintings and take a better look at it \n3. You reach for the pan in front of you")
            # User types option
            user_input = input("Choose an action (1, 2, or 3): ")
            # If user selects option 1
            if user_input == "1":
                # Dialogue and dialogue pauses
                print("The girl lets out a shriek and grabs the pan next to her.")
                sleep(1)
                print("\nRapunzel: 'Who are you guys and how did you get into my house?', she says.")
                sleep(1)
                # Referring to the username given by the user
                print(f"\nIndiana Jones:  'My name is Indiana Jones. It seems like my friend {self.name} and I have landed in your world.'")
                sleep(1)
                print(f"Indiana Jones: '{self.name} is helping me to return to my own world. What is your name?'")
                sleep(1)
                print("\nRapunzel: 'My name is Rapunzel.'")
                sleep(1)
                print("\nIndiana Jones: 'Well, nice to meet you Rapunzel. Could you maybe get us out of here?'")
                # Calls the next scene 
                self.escaping_tower()
            # If user selects option 2
            elif user_input == "2":
                # Dialogue and dialogue pauses
                print("The wooden floor creaks as you begin to move.")
                sleep(1)
                print("The girl turns towards the noise. She grabs the pan behind her. \nRapunzel: 'Who are you guys and how did you get in here?'")
                sleep(1)
                # Referring to the username given by the user
                print(f"\nIndiana Jones: 'Sorry, we didn't mean to scare you. I'm Indiana Jones and this is {self.name}.'")
                sleep(1)
                print("\nIndiana Jones: 'It seems like we have landed in your world.'")
                sleep(1)
                print(f"Indiana Jones: '{self.name} is helping me to return to my own world. What is your name?'")
                sleep(1)
                print("\nRapunzel: 'My name is Rapunzel.'")
                sleep(1)
                print("\nIndiana Jones: 'Well, nice to meet you Rapunzel. Could you maybe get us out of here?'")
                # Calls the next scene 
                self.escaping_tower()
            # If user selects option 3
            elif user_input == "3":
                # Dialogue and dialogue pauses
                print("You stumble over a loose wooden tile. She turns towards the noise. \nRapunzel: 'Who are you guys and what are you doing in my home?'")
                sleep(1)
                # Referring to the username given by the user
                print(f"\n{self.name}:  'My name is {self.name}. It seems like Indiana Jones and I have landed in your world.'")
                sleep(1)
                print(f"Indiana Jones: 'Yes, {self.name} is helping me to return to my own world. What is your name?'")
                sleep(1)
                print("\nRapunzel: 'My name is Rapunzel.'")
                sleep(1)
                print("\nIndiana Jones: 'Well, nice to meet you Rapunzel. Could you maybe get us out of here?'")
                # Calls the next scene 
                self.escaping_tower()
            # If input of the user isn't contained in the options, user should enter another option
            else:
                print("Please enter a valid option for this game.")

    # Escaping tower scene
    def escaping_tower(self):
        # Dialogue and dialogue pauses
        sleep(2)
        print("\nRapunzel: 'Frankly, I've never left this tower... I'm not sure if I can help you guys get out, but maybe together we can figure it out.'")
        sleep(1)
        print("\nYou look around the room. There are a few things that stand out: \n1. A narrow window high up \n2. A golden chest \n3. Rapunzels long hair")
        # List of options
        options = ["1", "2", "3"]
        # Pre-define variable user input
        user_input = ""
        # Loop until the user enters one of the given options
        while user_input not in options:
            # Dialogue and dialogue pauses
            sleep(2)
            print("\nWhat do you want to do?")
            print("1. Try climbing out of the window \n2. Open the golden chest \n3. Ask if Rapunzel could make a rope out of her hair")
            # User types option
            user_input = input("Choose an option (1, 2 or 3): ")
            # If user selects option 1
            if user_input == "1":
                # Calls the falling scene
                self.falling()
            # If user selects option 2
            elif user_input == "2":
                sleep(1)
                print("The golden chest is locked. You'll need to find the key first.")
                sleep(1)
                print("Rapunzel: 'Mother says I'm not allowed to open the chest.'")
                # Calls the key finding scene
                self.keyfinding()
            # If user selects option 3
            elif user_input == "3":
                self.option3_escaping_tower()
            # If input of the user isn't contained in the options, user should enter another option
            else:
                print("Please enter a valid option for this game.")
    
    # To avoid repetition of code this function is created to cover an option of escaping the tower
    def option3_escaping_tower(self):
        # Dialogue and dialogue pauses
        sleep(1)
        print("Of course! You know, I've always wanted to see the floating latterns at the castle. Could I maybe come along with you guys?")
        sleep(1)
        print("Indiana Jones: 'Of course you can come with us!'")
        sleep(1)
        print("As you step towards the edge of the tower, Rapunzel gathers her long, golden hair, securing it around a sturdy stone pillar. She tosses her just braided hair over the side.")
        sleep(1)
        print("You grab her braid and look down. It is definitely a pretty high tower. The wind whips your face. You are scared.")
        sleep(1)
        # List of possible answers
        answers = ["y", "n"]
        # Pre-define answer user will put in
        answer = ""
        # Loop until the user enters one of the given answers
        while answer not in answers:
            # User types answer
            answer = input("Rapunzel: 'Do you want us to count down before you jump?' (y/n)")
            # If answer is yes
            if answer == "y":
                # Dialogue and dialogue pauses
                sleep(1)
                print("You nod your head.")
                sleep(1)
                print("Rapunzel: 'OK, I'll count to three.'")
                sleep(1)
                # For-loop to count from 1 to 3
                for i in range(1, 4):
                    print(i)
                    # Dialogue and referring to user name
                    print(f"{self.name}: Here I go!")
                    # Calls the next scene
                    self.outsidetower_scene()
            # If answer is no
            elif answer == "n":
                # Dialogue and dialogue pause
                sleep(1)
                print("You hold your breath and jump, while holding onto the braided hair.")
                # Calls the next scene
                self.outsidetower_scene()
            # If input of the user isn't contained in the answer, user should enter another answer
            else:
                print("Please enter a valid option for this game.")

    # Falling scene
    def falling(self):
        # Dialogue and dialogue pauses
        sleep(2)
        print("You climb the wall to reach the window.")
        # Long dramatic pause
        sleep(3)
        print("Oh no, your foot slips of the steep wall. You fall. You lose 25 points.")
        sleep(1)
        # User loses 25 points 
        self.points = self.points - 25
        # Tells user how much points he has at the moment
        print(f"You've got {self.points} points.")
        sleep(1)
        print("You still want to escape. What will you do next?")
        # List of options
        options = ["1", "2"]
        # Pre-define variable user input
        user_input = ""
        # Loop until the user enters one of the given options
        while user_input not in options:
            # Dialogue and dialogue pauses
            sleep(2)
            print("\nWhat do you want to do?")
            sleep(1)
            print("1. Open the golden chest \n2. Ask if Rapunzel could make a rope out of her hair")
            # User types option
            user_input = input("Choose an option (1 or 2): ")
            sleep(1)
            # If user selects option 1
            if user_input == "1":
                # If key is already in the items list
                if self.key in self.items:
                    # Skips key finding scene and calls unlocking chest scene
                    self.unlockchest()
                # If user hasn't found the key yet
                else:
                    # Dialogue and dialogue pauses
                    sleep(1)
                    print("The golden chest is locked. You'll need to find the key first.")
                    sleep(1)
                    print("Rapunzel: 'Mother says I'm not allowed to open the chest.'")
                    # Calls the key finding scene
                    self.keyfinding()
            # If user selects option 2
            elif user_input == "2":
                # To avoid repetition, refers to escaping tower option
                self.option3_escaping_tower()
            # If input of the user isn't contained in the options, user should enter another option
            else:
                print("Please enter a valid option for this game.")

    # Finding the key scene
    def keyfinding(self):
        # Dialogue and dialogue pauses
        sleep(2)
        print("\nYou look around. Where will this key be?")
        sleep(1)
        # Prints three possible options
        print("\n1. In the drawer of the kitchen cabinet \n2. Underneath the loose wooden tile \n3. Underneath the plant pot")
        sleep(1)
        # Option where the key is hidden
        options = ["1"]
        # Pre-define variable user input
        user_input = ""
        # loop until the user enters the right option
        while user_input not in options:
            # User types option
            user_input = input("Where do you think the key will be? (1, 2 or 3)")
            # If input is option 1
            if user_input == "1":
                # Dialogue
                print("You've found the key. You unlock the chest with the key.")
                # Add key to item list
                self.items.append(self.key)
                # Calls the next scene
                self.unlockchest()
            # Prompts user to try a different option
            else:
                print("The key is not there. Please try again.")

    # Unlocking the chest scene
    def unlockchest(self):
        # If satchel is already in the items list
        if self.satchel in self.items:
            # Dialogue
            print("Looks like someone has already taken the stolen satchel and put it in his bag.")
        # If satchel is not in items list
        else:
            # Dialogue and dialogue pauses
            sleep(2)
            print("\nYou find the stolen satchel of the queen.")
            # List of answers
            answers = ["y", "n"]
            # Pre-define variable user input
            answer = ""
            # Loop until the user enters one of the given answers
            while answer not in answers:
                # User types answer
                answer = input("Do you want to put the satchel in your bag? (y/n)")
                # If answer is yes
                if answer == "y":
                    # Add satchel to items list
                    self.items.append(self.satchel)
                    print("You've succesfully put the satchel in your bag.")
                # If answer is no
                elif answer == "n":
                    print("Indiana Jones: 'Well, I guesss the Queen won't miss it.'")
                # If input of the user isn't contained in the options, user should enter another answer
                else:
                    print("Please enter a valid option.")
        # Dialogue and dialogue pause
        sleep(1)
        print("\nRapunzel: 'You guys really have to go before mother comes.'")
        # List of options
        options = ["1", "2"]
        # Pre-define variable user input
        user_input = ""
        # Loop until the user enters one of the given options
        while user_input not in options:
            print("1. Try climbing out of the window \n2. Ask if Rapunzel could make a rope out of her hair")
            # User enters option
            user_input = input("What do you want to do?")
            # If input is option 1
            if user_input == "1":
                # Calls falling scene
                self.falling()
            # If input is option 2
            elif user_input == "2":
                # Dialogue and dialogue pause
                sleep(1)
                print("Of course! You know, I've always wanted to see the floating latterns at the castle. Could I maybe come along with you guys?")
                print("Indiana Jones: 'Of course you can come with us!'")
                # Calls outside the tower scene
                self.outsidetower_scene()
            # If input of the user isn't contained in the options, user should enter another answer
            else:
                print("Please enter a valid option for this game.")

    # Scene outside the tower
    def outsidetower_scene(self):
        # Dialogue pause
        sleep(2)
        # User gains 25 points
        self.points = self.points + 25
        # Tells how much points user currently has
        print(f"\nYou've helped Rapunzel escaped the tower. You gain 25 points. You've got {self.points} points.")
        # Dialogue and dialogue pauses
        sleep(1)
        print("Help Rapunzel get to the castle to see the floating lights.")
        sleep(1)
        print("\nThe are surrounded by trees. Looking west, you see the castle rising above the trees.")
        sleep(1)
        print("You see a beautiful white horse that seems to be stuck in a bush.")
        sleep(1)
        print("There's not a lot of time before the mother of Rapunzel returns, what do you do?")
        print("\n1. There's no time, you start walking towards the castle \n2. You free the horse")
        sleep(1)
        # List of options
        options = ["1", "2"]
        # Pre-define variable user input
        user_input = ""
        # Loop until the user enters one of the given options
        while user_input not in options:
            # User enters option
            user_input = input("What would you like to do? (1 or 2)")
            # If input is option 1
            if user_input == "1":
                # Dialogue
                print("The three of you start heading west straight through the forest.")
            # If input is option 2
            elif user_input == "2":
                # Dialogue and dialogue pause
                print("The horse thanks you by licking your face.")
                sleep(1)
                # User gains 10 points
                self.points = self.points + 10
                # Tells how much points user currently has 
                print(f"You gain 10 points for doing a good deed. You've got {self.points} points.")
                print("The horse happily drafts away. You continue heading west.")
            # If input of the user isn't contained in the options, user should enter another answer
            else:
                print("Please enter a valid option for this game.")
        # Dialogue and dialogue pauses
        sleep(1)
        print("While heading your way towards the castle, you hear leaves cracking behind you.")
        sleep(1)
        print("You turn towards the noise and see an old, mean looking lady hiding half behind the trees. \nBehind her there are two enormous fully-inked men. \nThey look like twins. One of the men is wearing an eyepatch. \nShe's looking at Rapunzel.")
        sleep(2)
        print("Mother Gothel: 'Rapunzel, what did I tell you about talking to strangers? What are you doing with these people?'")
        sleep(1)
        print("Rapunzel: 'I went with them to see the floating latterns, mother.'")
        sleep(1)
        print("Mother Gothel: 'Don't be silly Rapunzel, come with me and we'll go home.'")
        sleep(1)
        print("Rapunzel: 'Well, I don't want to come with you.'")
        sleep(1)
        print("Mother Gothel tugs on Rapunzel her arm. \nMother Gothel: 'You need to come with me child.'")
        sleep(1)
        print("Rapunzel: 'Let go of me.'")
        sleep(1)
        print("Indiana Jones: 'You heard the lady. Let her go.'")
        sleep(1)
        print("Mother Gothel: 'I am not letting her go without a fight.'")
        sleep(1)
        print("The men who were once behind her are now standing infront of you.")
        sleep(1)
        print("Indiana Jones looks towards you and winks. \nIndiana Jones: 'Well, then a fight you shall have.'")
        sleep(1)
        print("Indiana Jones: 'You take one and I take the other.'")
        sleep(1)
        print("The man that's approaching you has a sword.")
        sleep(1)
        print("You look around.")
        sleep(2)
        print("Rapunzel has a pan attached to her dress. There is also a sword lying on the ground.")
        sleep(1)
        # Calls choosing weapon function
        self.choose_weapon()

    def choose_weapon(self):
        # Makes sure that health of user is 100 (full health) before each fight
        you.health = 100
        # Health of enemy is 100 (full health)
        twinbrother.health = 100
        # List of options
        options = ["fists", "pan", "sword"]
        # Pre-define variable user input
        user_input = ""
        # Loop until the user enters one of the given options
        while user_input not in options:
            # User enters option
            user_input = input("Choose your own weapon: fists/pan/sword")
            # If input is fists
            if user_input == "fists":
                # Calls equip function to equip user with fists as weapon
                you.equip(fists)
                # Calls fighting game
                self.attack()
            # If input is pan
            elif user_input == "pan":
                # Calls equip function within character file to equip users' input weapon
                you.equip(pan)
                # Calls fighting game
                self.attack()
            # If input is sword
            elif user_input == "sword":
                # Calls equip function within character file to equip users' input weapon
                you.equip(sword)
                # Calls fighting game 
                self.attack()
            # If input of the user isn't contained in the options, user should enter another answer
            else: 
                print("Please enter a valid option.")

    # Fighting scene
    def attack(self):
        # Clears the terminal screen
        os.system("cls")
        # Dialogue pause
        sleep(2)
        # Starts an infinite loop for the fight until user loses/wins
        while True:
            # User presses Enter
            enter = input("Press Enter to attack")
            # If user presses enter
            if enter == "":
                # User attack twin brother
                you.attack(twinbrother)
                # Twin brother attacks user
                twinbrother.attack(you)
                # Draw users health bar
                you.health_bar.draw()
                # Draw twin brothers' health bar
                twinbrother.health_bar.draw()
                
                # If user loses (meaning users' health = 0)
                if you.health == 0:
                    # Dialogue and dialogue pause
                    sleep(2)
                    print("Sadly, you lost the fight. Please try again with a different weapon.")
                    # Rematch, possibily choosing another weapon
                    self.choose_weapon()
                # If users'health is not zero
                else:
                    # If user wins (meaning twin brothers' health = 0)
                    if twinbrother.health == 0:
                        # User gains 25 points
                        self.points = self.points + 25
                        # Dialogue and dialogue pauses
                        sleep(2)
                        # Tells user how much points he/she has currently got
                        print(f"Hoora, you've beated the men with the eye patch. \nYou gain 25 points. You've got {self.points} points.")
                        sleep(1)
                        print("You look to your right and see that Indiana Jones has beaten the other twin brother.")
                        sleep(1)
                        print(f"Rapunzel grabs your hand. \nRapunzel: 'We have to go now {self.name}!'")
                        sleep(1)
                        print("You run towards the village.")
                        # Calls the next scene
                        self.village()
                    # Continue pressing enter
                    else:
                        print("Press Enter to attack")
            # If input of the user isn't Enter, user should enter Enter
            else: 
                print("Please press Enter to attack.")

    # Village scene
    def village(self):
        # Dialogue and dialogue pauses
        sleep(3)
        print("You enter the lively village, where music fills the air and people dance joyfully. \nLanterns sway above as vendors sell fresh pastries and flowers. \nChildren laugh, weaving through the crowd. \nThe castle stands tall in the distance.")
        sleep(1)
        print("While you look around, a kid hops towards you and gives you a flower.")
        sleep(1)
        print("What do you do? \n1. Dance with the villagers \n2. Go to the castle \n3. Introduce yourself to the kid and tell a joke")
        sleep(1)
        # List of options
        options = ["1", "2", "3"]
        # Pre-define variable user input
        user_input = ""
        # Loop until the user enters one of the given options
        while user_input not in options:
            # User enters option
            user_input = input("Choose an option (1, 2 or 3): ")
            sleep(1)
            # If input is option 1
            if user_input == "1":
                # User gains 10 pints
                self.points = self.points + 10
                # Dialogue and dialogue pauses
                print(f"Way to go! An adventure isn't an adventure without a little fun. \nDue to the happiness of your travelcompanions gain 10 points. You've got {self.points} points.")
                sleep(3)
                print("After dancing you make your way towards the castle.")
                # Calls the castle function
                self.castle()
            # If input is option 2
            elif user_input == "2":
                # Dialogue
                print("You make your way towards the castle.")
                # Calls the castle function
                self.castle()
            # If input is option 3
            elif user_input == "3":
                # Calls the joke telling function
                self.joke_telling()
            # If input of the user isn't contained in the options, user should enter another answer
            else:
                print("Please enter a valid option for this game.")

    # Function to tell a joke
    def joke_telling(self):
        #Two possible joke questions and corresponding answers separated by commas 
        jokes = (
            (f"\n{self.name}:'Whats powerful enough to clear any obstacle from Rapunzels path?'", f"n{self.name}: 'A hair-icane!'"),
            (f"\n{self.name}:'Why did Rapunzel never experience fun?'", f"{self.name}: 'She never let her hair down.'")
        )
        # Two possible reactions of the kid, result of the reaction and change in points of user according to the reaction of the kid
        reaction_kid = (
            ("\nThe kid likes the joke and starts to laugh.", f"You gain ten points. You've got {self.points} points.", self.points + 10),
            ("\nThe joke wasn't that funny. The kid doesn't laugh.", f"You lose ten points. You've got {self.points} points.", self.points - 10)
        )
        # Tell a random joke
        joke, answer = choice(jokes)
        # First print the joke
        print(joke)
        # Wait for answer
        sleep(2)
        # Then print the answer of the joke
        print(answer)
        # Wait for reaction
        sleep(2)
        # Kids reaction (random)
        reaction, result, change_points = random.choice(reaction_kid)
        # Print the reaction of the kid
        print(reaction)
        # Adjust users points according to the reaction
        self.points += change_points 
        # Print the result of the reaction of the kid
        print(result)
        # Dialogue and dialogue pause
        print("You wave the kid goodbye and make your way towards the castle.")
        sleep(1)
        # Calls the next scene
        self.castle()

    # Castle scene
    def castle(self):
        # Dialogue and dialogue pauses
        print("You passed the guards and the grand door creaks open. \nThe door opens to a grand hall surrounded with beautiful paintings.")
        sleep(1)
        print("At the end of the hall the queen and king sit on the throne.")
        # If satchel is contained in list of items
        if self.satchel in self.items:
            # Dialogue and dialogue pauses
            print("Indiana Jones: 'Your majesty, we've brought you your stolen satchel.'")
            sleep(1)
            print("You get the satchel out of your bag and bring it towards them.")
            sleep(1)
            # Tells how much points user gained and currently has
            print(f"Hoora, you've returned the satchel to the queen. You gain 25 points. You've got {self.points} points.")
        # If satchel isn't contained in list of items
        else:
            print("Queen: 'Come in, what is it you want?'")
        # Dialogue and dialogue pauses
        sleep(1)
        print("The queens' gaze locks on Rapunzel. \nSlowly, she stands, her hand covering her mouth, eyes glistening.")
        sleep(1)
        print("Queen: '...Rapunzel?'")
        sleep(1)
        print("Rapunzel stiffens and looks confused.")
        sleep(1)
        print("Queen: it's you... my daughter. You were stolen many years ago by a witch.")
        sleep(1)
        print("Rapunzel's breath catches. \nEmotion overwhelms her as she collapses into the Queen's embrace.")
        sleep(1)
        print("Indiana Jones glances at you, raising an eyebrow.")
        sleep(1)
        print("Guess the stolen satchel wasn't the real treasure after all.")
        # User gains 40 points
        self.points = self.points + 40
        # Tells how much points user gained and currently has
        print(f"Hoorah, you've returned the long lost daughter of the queen. \nYou gain 40 points. \nYou've got {self.points} points.")
        sleep(1)
        # If user has less than 100 points achieved in this world
        if self.points < 100:
            # Dialogue
            print("For you this adventure has not come to an end, since you have not reached 100 points. \nIn order for you to complete this world and gain a portal piece, you'll need 100 points.")
            # List of answers
            answers = ["leave","retry"]
            # Pre-define variable answer
            answer = ""
            # Loop until the user enters one of the given answers
            while answer not in answers:
                # User types answer
                answer = input("Do you want to leave this world without getting a portal piece or try again? (leave/retry)")
                # If input is leave
                if answer == "leave":
                    print("You've not gained a portal piece and will be transported back to the portal room.")
                    # Go back to portal choosing scene
                    self.portalscene()
                # If input is retry
                elif answer == "retry":
                    sleep(2)
                    print("You'll transport to the beginning of this adventure to try again. \nPick a different path to earn more points.")
                    self.first_world()
                # If input of the user isn't contained in the options, user should enter another answer
                else:
                    print("Please enter a valid option.")
        # If user has gained 100 points or more in this world
        else:
            # User gains the portal piece of first world
            self.portalpieces.append(self.first_portal_piece)
            # Tells which portal pieces user currently has
            print(f"Hoorah, you've completed this world and achieved a portal piece! \nYou've got the following portal pieces: {self.portalpieces}.")
            # Calls the portal scene
            self.portalscene()

    
    # --------------- Second world:  ---------------
    
    def second_world(self):
        # If the user has already achieved the second portal piece
        if self.second_portal_piece in self.portalpieces:
            # Pre-define user variable
            user_input = ""
            # List of answers
            answers = ["y", "n"]
            # Loop until the user enters one of the given options
            while user_input not in answers:
                # User types answer
                user_input = input("You've already completed this world. If you start again, you'll need to earn this portal piece again. Do you want to continue? (y/n)")
                # If answer is yes
                if user_input == "y":
                    # Remove the second portal piece from the list of portal pieces
                    self.portalpieces.remove(self.second_portal_piece)
                # If answer is no
                elif user_input == "n":
                    # User goes back to portal scene
                    self.portalscene()
                # If input of the user isn't contained in the options, user should enter another answer
                else:
                    print("Please enter a valid option.")
        # begin elke wereld met code regel:
        self.points = 0
        # voeg punten toe bv. door:
        # self.points = self.points + 25
        # als je alle punten behaald hebt, krijg je een portal piece door:
        if self.points > 100:
            # voegt portal piece van jouw wereld toe aan lijst van portal pieces
            self.portalpieces.append(self.second_portal_piece)
            # ga na het afronden van een wereld terug naar self.portalscene() door:
            self.portalscene()
        else:
            # terug naar begin van je wereld? of kies zelf een optie
            self.second_world()    

    
    # --------------- Third world: ---------------
    def third_world(self):
        # If the user has already achieved the third portal piece
        if self.third_portal_piece in self.portalpieces:
            # Pre-define user variable
            user_input = ""
            # List of answers
            answers = ["y", "n"]
            # Loop until the user enters one of the given options
            while user_input not in answers:
                # User types answer
                user_input = input("You've already completed this world. If you start again, you'll need to earn this portal piece again. Do you want to continue? (y/n)")
                # If answer is yes
                if user_input == "y":
                    # Remove the third portal piece from the list of portal pieces
                    self.portalpieces.remove(self.third_portal_piece)
                # If answer is no
                elif user_input == "n":
                    # User goes back to portal scene
                    self.portalscene()
                # If input of the user isn't contained in the options, user should enter another answer
                else:
                    print("Please enter a valid option.")
        # begin elke wereld met code regel:
        self.points = 0
        # voeg punten toe bv. door:
        # self.points = self.points + 25
        # als je alle punten behaald hebt, krijg je een portal piece door:
        if self.points > 100:
            # voegt portal piece van jouw wereld toe aan lijst van portal pieces
            self.portalpieces.append(self.third_portal_piece)
            # ga na het afronden van een wereld terug naar self.portalscene() door:
            self.portalscene()
        else:
            # terug naar begin van je wereld? of kies zelf een optie
            self.third_world()
    
    # --------------- Fourth world: ---------------
    
    def fourth_world(self):
        # If the user has already achieved the fourth portal piece
        if self.fourth_portal_piece in self.portalpieces:
            # Pre-define user variable
            user_input = ""
            # List of answers
            answers = ["y", "n"]
            # Loop until the user enters one of the given options
            while user_input not in answers:
                # User types answer
                user_input = input("You've already completed this world. If you start again, you'll need to earn this portal piece again. Do you want to continue? (y/n)")
                # If answer is yes
                if user_input == "y":
                    # Remove the fourth portal piece from the list of portal pieces
                    self.portalpieces.remove(self.fourth_portal_piece)
                # If answer is no
                elif user_input == "n":
                    # User goes back to portal scene
                    self.portalscene()
                # If input of the user isn't contained in the options, user should enter another answer
                else:
                    print("Please enter a valid option.")

        print("Welcome to the Jumanji world! You have been transported into the an exciting game where you have to find balance between risk and reward.")
        print(f"\nHello {self.name}, you are now inside the world of Jumanji! In order to win this game you'll need to earn 15 points.")
        print("Your goal is to return alive with the Jaguar's Eye to the statue before Van Pelt stops you!")
        print("Be very cautious! You only have three lives, and every choice shapes your destiny.\n")
    
        # Adventure Points and Lives
        adventure_points = 0
        lives = 3
        required_points = 15  # Adjusted to match the new balancing
    
        # First Decision: Avatar Abilities
        print("Choose your special skill:")
        print("1. Strength and Speed (Dr. Smolder Bravestone)")
        print("2. Stealth and Agility (Ruby Roundhouse)")
        print("3. Zoology and Weapons (Mouse Finbar)")
        print("4. Cartography and Knowledge (Professor Oberon)")
    
        while True:
            try:
                choice = int(input("Enter the number of your chosen skill (1-4): "))
                if choice in [1, 2, 3, 4]:
                    break
                print("Invalid choice! Please enter a number between 1 and 4.")
            except ValueError:
                print("Invalid input! Please enter a valid number.")
    
        print("\nYou find yourself in the jungle, greeted by Nigel, an NPC who gives you a brief on your quest.")
        print("Nigel hands you the Jaguar's Eye, warning you that Van Pelt is hunting you.")
    
        # First Challenge: Path Selection
        print("\nYou arrive at a fork in the path. What will you do?")
        print("1. Venture into the dark jungle")  # cost a life but rewarding
        print("2. Walk along the riverbank")  # safer but lower points
        print("3. Climb up the tall trees to scout the area")  # safer but lower points
    
        while True:
            try:
                choice = int(input("Enter the number (1-3): "))
                if choice == 1:
                    adventure_points += 7
                    lives -= 1
                elif choice == 2:
                    adventure_points += 4
                elif choice == 3:
                    adventure_points += 3
                else:
                    print("Invalid choice! Please enter a number between 1 and 3.")
                    continue
                break
            except ValueError:
                print("Invalid input! Please enter a valid number.")

        print("You had your first major challenge accomplished! Keep up the good work!")
    
        # Second Decision: River or Cave
        print("\nYou reach a wide river with strong rapids and see a dark cave nearby. Do you:")
        print("1. Brave the rapids using an old raft.")  # cost a life but rewarding
        print("2. Enter the cave and find another route.")  # safer but lower points
        print("3. Look for an old bridge crossing.")  # safer but lower points
    
        while True:
            try:
                choice = int(input("Enter 1, 2, or 3: "))
                if choice == 1:
                    adventure_points += 6
                    lives -= 1
                elif choice == 2:
                    adventure_points += 3
                elif choice == 3:
                    adventure_points += 2
                else:
                    print("Invalid choice! Please enter 1, 2, or 3.")
                    continue
                break
            except ValueError:
                print("Invalid input! Please enter a valid number.")

        # Third Decision: The Ruins Puzzle
        print("\nYou stumble upon the ruins of an ancient temple. To proceed, you must find the right way through.")
        print("1. Solve an ancient puzzle carved into the stone.")  # safer but lower points
        print("2. Try to force open a hidden door with brute strength.")  # costs a life but rewarding
        print("3. Search for a secret passage in the overgrown vines.")  # safer but lower points

        while True:
            try:
                choice = int(input("Enter 1, 2, or 3: "))
                if choice == 1:
                    adventure_points += 2
                elif choice == 2:
                    adventure_points += 5
                    lives -= 1
                elif choice == 3:
                    adventure_points += 3
                else:
                    print("Invalid choice! Please enter 1, 2, or 3.")
                    continue
                break
            except ValueError:
                print("Invalid input! Please enter a valid number.")

        # Fourth decision: Reaching the Jaguar Statue
        print("\nAt last, you reach the Jaguar Statue, but Van Pelt and his men guard it fiercely!")
        print("How will you proceed?")
        print("1. Create a diversion using a thrown rock.")  # cost a life but rewarding
        print("2. Stealthily maneuver around and place the jewel unnoticed.")  # safer but lower points
    
        while True:
            try:
                choice = int(input("Enter 1 or 2: "))
                if choice == 1:
                    adventure_points += 4
                    lives -= 1
                elif choice == 2:
                    adventure_points += 2
                else:
                    print("Invalid choice! Please enter 1 or 2.")
                    continue
                break
            except ValueError:
                print("Invalid input! Please enter a valid number.")
    
        # Conclusion
        print(f"\nYou're finished, {self.name}! Let's see the results!")
        print(f"Your final adventure score: {adventure_points} points.")
        print(f"Lives remaining: {lives}")
    
        if adventure_points >= required_points and lives > 0:
            print("You made it out alive and returned to reality! Well done! You took risks when necessary and remained cautious when needed!")
            # User earns portal piece of this world
            self.portalpieces.append(self.fourth_portal_piece)
            print("You've won the portal piece of this world. Congrats!")
        else:
            print("You have not met the required points or ran out of lives... You remain trapped in Jumanji! Tip: find a better balance between risk and reward, choose another avatar to get other strenghts.")
            # Goes back to beginning of this world
            self.fourth_world()
        while True:
            restart_game = input("Would you like to try again? (yes/no): ").strip().lower()
            if restart_game == "yes":
                # Go back to the beginning of this world
                self.fourth_world()
                break
            elif restart_game == "no":
                print("Thanks for playing! See you next time.")
                # Go back to portal scene
                self.portalscene()
                break
            else:
                print("Invalid input! Please enter 'yes' or 'no'.")

    # --------------- Start game function ---------------
    def start_game(self):
        while not self.game_over:
            pass

# --------------- Start game ---------------
# Start the game (Player has 0 points when starting the game)
if __name__ == "__main__":
    game = movieportal_game(name="Player", points=0)
    game.start_game()
    

```
